##  DATA ANALYSIS

1. Are the number of reviews increasing each year?
2. Which book received the most number of reviews?

### Q1. Are the number of reviews increasing each year?

**Answer: The number of reviews have been moving up and down since 2012. There was a spike from 2017 to 2018.**

```python
y_reviews = best.groupby('Year', as_index=False).agg({'Reviews':'sum'})
y_reviews.head(11)
```

### Q2. Which book received the most number of reviews?

**Answer: The most reviewed book is "Where the Crawdads Sing" by Delia Owens.**

```python
best_reviews = best.sort_values('Reviews', ascending=False).head(10)
best_reviews
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.lineplot(x='Year', y='Reviews', data=y_reviews, marker='o')
plt.xlabel('')
plt.ylabel('No. of Reviews')
plt.show()
```

```python
sns.barplot(x='Reviews', y='Name', data=best_reviews, ci=False, orient='h', palette='pastel')
plt.title('Most Reviews Books')
plt.xlabel('')
plt.ylabel('')

plt.show()
```
